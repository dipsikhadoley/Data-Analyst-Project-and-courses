```python
import pandas as pd
import os
import logging
import time
from sqlalchemy import create_engine

logging.basicConfig(
    filename="logs/ingestion_db.log",
    level=logging.DEBUG,
    format="%(asctime)s - %(levelname)s - %(message)s",
    filemode="a"
)

# create engine once
engine = create_engine('sqlite:///inventory.db')

def ingest_db(df, table_name, engine):
    """Ingests dataframe into a database table."""
    df.to_sql(table_name, con=engine, if_exists='replace', index=False)
    logging.info(f"Table {table_name} ingested successfully with {len(df)} rows.")

def load_raw_data():
    """Loads all CSVs in 'data' folder as DataFrames and ingests into DB."""
    start = time.time()
    for file in os.listdir('data'):
        if file.endswith('.csv'):
            df = pd.read_csv(os.path.join('data', file))
            table_name = file[:-4]   # strip ".csv"
            logging.info(f"Ingesting {file} into table {table_name}...")
            ingest_db(df, table_name, engine)

    end = time.time()
    total_time = (end - start) / 60
    logging.info("------Ingestion Complete-------")
    logging.info(f"Total time taken: {total_time:.2f} minutes")

if __name__ == '__main__':
    load_raw_data()
```
